### From the heatmap we can say that there is multicollinearity within multiple variables
- Multicollinearity occurs when two or more predictor variables in a regression model are highly correlated, meaning they provide redundant information about the response variable. This can lead to several disadvantages, such as inflating the variance of the coefficient estimates, making them unstable and difficult to interpret. It also reduces the model's overall statistical power, leading to unreliable significance tests and potentially incorrect conclusions about the relationships between variables. Additionally, multicollinearity can impair the model's ability to generalize to new data, negatively affecting its predictive performance.

## Step 4: Removing Multicollinearity by calculating the Variance Inflation Factor (VIF) for each feature.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from patsy import dmatrices
```

```python
X = df.drop(columns=['diagnosis'])
```

```python
def calculate_vif(X):
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif
```

```python
vif = calculate_vif(X)
print(vif)
```
